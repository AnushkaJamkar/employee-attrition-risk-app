```python
df = pd.read_csv("data/employee_attrition.csv", sep="\t")
df.head()

```

```python
df['Attrition'].value_counts()
```

```python
df.info()
```

```python
cols_to_drop = [
    'EmployeeCount',
    'EmployeeNumber',
    'Over18',
    'StandardHours'
]

df = df.drop(columns=cols_to_drop)
```

```python
df.columns
```

```python
df['Attrition'] = df['Attrition'].map({'No': 0, 'Yes': 1})
```

```python
df['Attrition'].value_counts()
```

```python
df.groupby('OverTime')['Attrition'].mean()
```

```python
df.groupby('JobSatisfaction')['Attrition'].mean()
```

```python
X = df.drop('Attrition', axis=1)
y = df['Attrition']
```

```python
X.shape, y.shape
```

```python
X_encoded = pd.get_dummies(X, drop_first=True)
```

```python
X_encoded.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression(max_iter=3000)
model_lr.fit(X_train, y_train)
```

```python
y_pred_lr = model_lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

confusion_matrix(y_test, y_pred_lr)
```

```python
print(classification_report(y_test, y_pred_lr))
```

The logistic regression model achieves high accuracy, but has very low recall for attrition. Since missing a high-risk employee is costly, this model is not suitable for decision-making and serves only as a baseline.

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight='balanced'
)

model_rf.fit(X_train, y_train)
```

```python
y_pred_rf = model_rf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

confusion_matrix(y_test, y_pred_rf)
```

```python
print(classification_report(y_test, y_pred_rf))
```

A more complex model does not automatically give better business outcomes. Model evaluation must align with the cost of errors

```python
y_prob_rf = model_rf.predict_proba(X_test)[:, 1]
```

```python
import numpy as np

# try a lower threshold
threshold = 0.3

y_pred_rf_03 = (y_prob_rf >= threshold).astype(int)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

confusion_matrix(y_test, y_pred_rf_03)
```

```python
print(classification_report(y_test, y_pred_rf_03))
```

We deliberately lowered the decision threshold to improve recall for attrition, prioritizing business cost over raw accuracy

```python
results = X_test.copy()
results['Actual_Attrition'] = y_test.values
results['Attrition_Probability'] = y_prob_rf
```

```python
def risk_bucket(p):
    if p >= 0.6:
        return 'High'
    elif p >= 0.3:
        return 'Medium'
    else:
        return 'Low'

results['Risk_Level'] = results['Attrition_Probability'].apply(risk_bucket)
```
